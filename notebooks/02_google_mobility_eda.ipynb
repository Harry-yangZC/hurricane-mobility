# 02 — Mobility EDA: timeseries and heatmaps

This notebook loads the cleaned Google Community Mobility Parquet (produced by NB01) and generates a comprehensive set of visualisations:
1. National daily + 7-day rolling timeseries
2. Per-state small multiples
3. State-by-week heatmaps
4. Holiday-annotated timeseries
5. National & state day-of-week profiles
6. Hurricane event-window plots
7. Year-over-year landfall-state comparisons

All figures are saved to `output/figures/`.

```python
from hurricane_mobility.config import PROCESSED_DIR, FIGURES_DIR, ensure_dirs
from hurricane_mobility.lookups import GOOGLE_METRICS, CENSUS_REGIONS
from hurricane_mobility import plotting

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 140)

ensure_dirs()

METRICS = GOOGLE_METRICS

# Load processed Parquet
DATA_PATH = PROCESSED_DIR / 'us_mobility_2020_2022.parquet'
print(f'Loading {DATA_PATH}')
df = pd.read_parquet(DATA_PATH)
print(df.shape)

# Geographic level flags
is_national = df['sub_region_1'].isna()
is_state = df['sub_region_1'].notna() & df['sub_region_2'].isna()

df_national = df.loc[is_national].copy()
df_state = df.loc[is_state].copy()

df['date'] = pd.to_datetime(df['date'])
df_national['date'] = pd.to_datetime(df_national['date'])
df_state['date'] = pd.to_datetime(df_state['date'])

print({'national': len(df_national), 'state': len(df_state)})
```

```python
# National timeseries: daily and 7-day rolling means
plotting.plot_national_timeseries(
    df_national, METRICS,
    save_path=FIGURES_DIR / 'national_timeseries.png',
)
```

```python
# Per-state small multiples for selected metrics
plotting.plot_state_small_multiples(
    df_state, METRICS,
    save_path_prefix=str(FIGURES_DIR / 'state_sm_'),
)
```

```python
# State-by-week heatmaps
plotting.plot_state_week_heatmaps(
    df_state, METRICS, CENSUS_REGIONS,
    save_path=FIGURES_DIR / 'state_week_heatmaps.png',
)
```

```python
# National timeseries with selected US holiday labels
holidays = plotting.selected_holidays(years=(2020, 2021, 2022))
plotting.plot_national_timeseries_with_holidays(
    df_national, METRICS, holidays,
    save_path=FIGURES_DIR / 'national_holidays.png',
)
```

```python
# National day-of-week (DOW) profiles by year
plotting.plot_national_dow_profiles(
    df_national, METRICS,
    save_path_prefix=str(FIGURES_DIR / 'national_dow_'),
)
```

```python
# State-level DOW profiles for selected states
sel_states = ['Louisiana', 'Florida', 'Massachusetts', 'Colorado', 'California']
plotting.plot_state_dow_profiles(
    df_state, METRICS, sel_states,
    save_path_prefix=str(FIGURES_DIR / 'state_dow_'),
)
```

```python
# Hurricane event-window plots
EVENTS = [
    {
        'name': 'Laura (LA)',
        'window_start': pd.Timestamp('2020-08-24'),
        'window_end': pd.Timestamp('2020-08-29'),
        'landfall': pd.Timestamp('2020-08-27'),
        'landfall_state': 'Louisiana',
    },
    {
        'name': 'Zeta (LA)',
        'window_start': pd.Timestamp('2020-10-26'),
        'window_end': pd.Timestamp('2020-10-30'),
        'landfall': pd.Timestamp('2020-10-28'),
        'landfall_state': 'Louisiana',
    },
    {
        'name': 'Ida (LA)',
        'window_start': pd.Timestamp('2021-08-26'),
        'window_end': pd.Timestamp('2021-09-04'),
        'landfall': pd.Timestamp('2021-08-29'),
        'landfall_state': 'Louisiana',
    },
    {
        'name': 'Ian (FL)',
        'window_start': pd.Timestamp('2022-09-23'),
        'window_end': pd.Timestamp('2022-10-02'),
        'landfall': pd.Timestamp('2022-09-28'),
        'landfall_state': 'Florida',
    },
]

LINE_SERIES = [
    ('National', None),
    ('Louisiana', 'Louisiana'),
    ('Florida', 'Florida'),
    ('Massachusetts', 'Massachusetts'),
    ('Colorado', 'Colorado'),
]

plotting.plot_google_event_windows(
    df_national, df_state, EVENTS, LINE_SERIES, METRICS,
    save_path_prefix=str(FIGURES_DIR / 'google_event_'),
)
```

```python
# Year-over-year comparison for landfall state
EVENTS_WITH_STATE = [
    {'name': 'Laura (LA, 2020)', 'window_start': pd.Timestamp('2020-08-24'), 'window_end': pd.Timestamp('2020-08-29'), 'landfall': pd.Timestamp('2020-08-27'), 'landfall_state': 'Louisiana'},
    {'name': 'Zeta (LA, 2020)', 'window_start': pd.Timestamp('2020-10-26'), 'window_end': pd.Timestamp('2020-10-30'), 'landfall': pd.Timestamp('2020-10-28'), 'landfall_state': 'Louisiana'},
    {'name': 'Ida (LA, 2021)', 'window_start': pd.Timestamp('2021-08-26'), 'window_end': pd.Timestamp('2021-09-04'), 'landfall': pd.Timestamp('2021-08-29'), 'landfall_state': 'Louisiana'},
    {'name': 'Ian (FL, 2022)', 'window_start': pd.Timestamp('2022-09-23'), 'window_end': pd.Timestamp('2022-10-02'), 'landfall': pd.Timestamp('2022-09-28'), 'landfall_state': 'Florida'},
]

plotting.plot_google_yoy_landfall(
    df_state, EVENTS_WITH_STATE, METRICS,
    save_path_prefix=str(FIGURES_DIR / 'google_yoy_'),
)
```

```python
# Missing-data checks for event windows and YoY landfall comparisons
def check_missing_event_windows(df_national, df_state, events, line_series, metrics):
    issues = []
    for ev in events:
        start = ev['window_start'] - pd.Timedelta(days=14)
        end = ev['window_end'] + pd.Timedelta(days=14)
        idx = pd.date_range(start, end, freq='D')
        for label, st_name in line_series:
            src = df_national if st_name is None else df_state[df_state['sub_region_1'] == st_name]
            for m in metrics:
                s = src.set_index('date')[m].reindex(idx)
                pct = s.isna().mean() * 100
                if pct > 50:
                    issues.append(f"{ev['name']} | {label} | {m}: {pct:.0f}% missing")
    if issues:
        print('\n'.join(issues))
    else:
        print('No severe missing-data issues.')

check_missing_event_windows(df_national, df_state, EVENTS, LINE_SERIES, METRICS)

def check_missing_yoy_windows(df_state, events_ws, metrics, years=(2020, 2021, 2022)):
    issues = []
    for ev in events_ws:
        st = ev['landfall_state']
        s_md = ev['window_start'].strftime('%m-%d')
        e_md = ev['window_end'].strftime('%m-%d')
        for y in years:
            try:
                idx = pd.date_range(f'{y}-{s_md}', f'{y}-{e_md}', freq='D')
            except Exception:
                continue
            s = df_state[df_state['sub_region_1'] == st].set_index('date')
            for m in metrics:
                pct = s[m].reindex(idx).isna().mean() * 100
                if pct > 50:
                    issues.append(f"{ev['name']} | {st} | {y} | {m}: {pct:.0f}% missing")
    if issues:
        print('\n'.join(issues))
    else:
        print('No severe missing-data issues in YoY windows.')

check_missing_yoy_windows(df_state, EVENTS_WITH_STATE, METRICS)

print('\nAll NB02 figures saved to:', FIGURES_DIR)
```
